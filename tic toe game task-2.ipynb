```python
import numpy as np

# Function to print the Tic Tac Toe board
def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

# Function to check for winning conditions
def check_win(board, player):
    # Check rows, columns, and diagonals
    for i in range(3):
        if all(board[i][j] == player for j in range(3)) or \
           all(board[j][i] == player for j in range(3)):
            return True
    if all(board[i][i] == player for i in range(3)) or \
       all(board[i][2-i] == player for i in range(3)):
        return True
    return False

# Function to make AI move using minimum matrix algorithm
def ai_move(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'O'
                if check_win(board, 'O'):
                    return
                board[i][j] = ' '
    # If no winning move, make a random move
    while True:
        i, j = np.random.randint(3), np.random.randint(3)
        if board[i][j] == ' ':
            board[i][j] = 'O'
            return

# Function to play the Tic Tac Toe game
def play_game():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    print("Welcome to Tic Tac Toe!")
    print_board(board)
    player = 'X'
    while True:
        if player == 'X':
            row, col = map(int, input("Enter row and column (1-3) for your move (e.g., '1 2'): ").split())
            row -= 1
            col -= 1
            if board[row][col] != ' ':
                print("Invalid move! Cell already occupied.")
                continue
            board[row][col] = 'X'
            if check_win(board, 'X'):
                print("Congratulations! You win!")
                break
        else:
            print("AI is making a move...")
            ai_move(board)
            if check_win(board, 'O'):
                print("AI wins! Better luck next time.")
                break
        print_board(board)
        # Check for a draw
        if all(board[i][j] != ' ' for i in range(3) for j in range(3)):
            print("It's a draw!")
            break
        # Switch player
        player = 'X' if player == 'O' else 'O'

# Play the game
play_game()
```
